```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("""
============================================================
ðŸ“˜ What is Convolution?
Convolution combines two signals to produce a third signal.
It shows how one signal modifies or filters another.
============================================================
""")
```

```python
# ---------------------------------------------------------------
# Example 1: Linear Convolution
# ---------------------------------------------------------------

# Define two input signals
x = [1, 2, 3, 4]
h = [1, 0, -1]

# Perform linear convolution using numpy
linear_conv = np.convolve(x, h)

# Display results
print("\nExample 1: Linear Convolution")
print("Input x[n]:", x)
print("Input h[n]:", h)
print("Linear Convolution Output:", linear_conv)

# Plot the result
plt.figure(figsize=(6, 3))
plt.stem(linear_conv)  # Removed use_line_collection
plt.title("Linear Convolution Output")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Example 2: Circular Convolution
x = [1, 2, 3, 4]
h = [1, 0, -1]

N = max(len(x), len(h))
x_padded = np.pad(x, (0, N - len(x)), mode='constant')
h_padded = np.pad(h, (0, N - len(h)), mode='constant')

X_fft = np.fft.fft(x_padded)
H_fft = np.fft.fft(h_padded)
circular_conv = np.fft.ifft(X_fft * H_fft)
circular_conv = np.real(circular_conv)

print("\nExample 2: Circular Convolution")
print("Padded x[n]:", x_padded)
print("Padded h[n]:", h_padded)
print("Circular Convolution Output:", circular_conv)

plt.figure(figsize=(6, 3))
plt.stem(circular_conv)
plt.title("Circular Convolution Output")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# Example 3: Cross-Correlation
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])

cross_corr = np.correlate(x, y, mode='full')

print("\nExample 3: Cross-Correlation")
print("Signal x:", x)
print("Signal y:", y)
print("Cross-Correlation Output:", cross_corr)

plt.figure(figsize=(6, 3))
plt.stem(cross_corr)
plt.title("Cross-Correlation between x[n] and y[n]")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid(True)
plt.show()
```

```python
# Example 4: Autocorrelation
x = np.array([1, 2, 3, 4, 5])

auto_corr = np.correlate(x, x, mode='full')

print("\nExample 4: Autocorrelation")
print("Signal x:", x)
print("Autocorrelation Output:", auto_corr)

plt.figure(figsize=(6, 3))
plt.stem(auto_corr)
plt.title("Autocorrelation of x[n]")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid(True)
plt.show()
```
