```python
print("\n--- Example 1: Class & Object ---")

class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def car_details(self):
        return f"Car: {self.brand}, Model: {self.model}"

# Taking input from user
brand = input("Enter car brand: ")
model = input("Enter car model: ")

print(f"You entered -> Brand: {brand}, Model: {model}")
my_car = Car(brand, model)
print("Output:", my_car.car_details())
```

```python
print("\n--- Example 2: Rectangle Area & Perimeter ---")

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# Taking input
w = int(input("Enter rectangle width: "))
h = int(input("Enter rectangle height: "))

print(f"You entered -> Width: {w}, Height: {h}")
rect = Rectangle(w, h)
print("Output -> Area:", rect.area())
print("Output -> Perimeter:", rect.perimeter())
```

```python
class BankAccount:
    def __init__(self, holder, balance):
        self.holder = holder
        self.__balance = balance    # private

    def deposit(self, amt):
        self.__balance += amt
        print(f"Deposited {amt}. New balance = {self.__balance}")

    def withdraw(self, amt):
        if amt <= self.__balance:
            self.__balance -= amt
            print(f"Withdrew {amt}. New balance = {self.__balance}")
        else:
            print("Not enough balance")

    def get_balance(self):
        return self.__balance


# --- Main program ---
name = input("\nEnter account holder name: ")
bal = float(input("Enter initial balance: "))
acc = BankAccount(name, bal)

print(f"Account created for {name} with balance {bal}")

d = float(input("Enter deposit amount: "))
acc.deposit(d)   # now prints inside method

w = float(input("Enter withdraw amount: "))
acc.withdraw(w)  # now prints inside method

print("Final balance is:", acc.get_balance())
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        return self.name + " says Woof!"

class Cat(Animal):
    def speak(self):
        return self.name + " says Meow!"

dog_name = input("\nEnter dog name: ")
cat_name = input("Enter cat name: ")

d = Dog(dog_name)
c = Cat(cat_name)

print(d.speak())
print(c.speak())
```

```python
class Polygon:
    def render(self):
        print("Drawing polygon")

class Square(Polygon):
    def render(self):
        print("Drawing square")

class Circle(Polygon):
    def render(self):
        print("Drawing circle")

print("\nPolymorphism example:")
shapes = [Square(), Circle(), Polygon()]
for s in shapes:
    s.render()
```

```python
from abc import ABC, abstractmethod

# abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# circle class
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r

# rectangle class
class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

# main
r = int(input("Enter radius of circle: "))
c = Circle(r)
print("Circle area =", c.area())

w = int(input("Enter rectangle width: "))
h = int(input("Enter rectangle height: "))
rect = Rectangle(w, h)
print("Rectangle area =", rect.area())
```
