```python
total = 0
for i in range(1, 101):
    total += i
print(total)
```

```python
numbers = [10, 21, 4, 45, 66, 93, 100]

even_count = 0
for n in numbers:
    if n % 2 == 0:
        even_count += 1

print("Number of even values:", even_count)
```

```python
text = "Hello World"
rev = ""

for ch in text:
    rev = ch + rev

print("Reversed string:", rev)
```

```python
numbers = [12, 45, 3, 67, 22, 90, 11]

largest = numbers[0]
for n in numbers:
    if n > largest:
        largest = n

print("Largest number:", largest)
```

```python
n = int(input("Enter how many terms: "))

a, b = 0, 1
print("Fibonacci series:")

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
items = [2, 4, 2, 8, 4, 9, 8, 1]

result = []
for i in items:
    if i not in result:
        result.append(i)

print("List after removing duplicates:", result)
```
