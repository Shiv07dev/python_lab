```python
import matplotlib.pyplot as plt

# simple dataset
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]

plt.plot(x_values, y_values, color="blue", marker='d', linestyle='--')
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.title("Line Graph Example")
plt.legend(["Speed vs Time"])
plt.grid(True, linestyle=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 2*np.pi, 40)
y = np.exp(np.sin(x))

plt.stem(x, y, linefmt="g-", markerfmt="ro", basefmt="grey")
plt.title("Stem Plot Example")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D", "E"]
y = [5, 7, 3, 8, 4]

plt.bar(x, y, color="skyblue", edgecolor="black")
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend(["Bars"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 4]

plt.hist(
    x, 
    bins=[1,2,3,4,5,6,7], 
    color="skyblue",       # softer color
    edgecolor="black", 
    alpha=0.8              # slight transparency
)
plt.title("Histogram Example", fontsize=14, fontweight="bold")
plt.xlabel("Bins", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(["Histogram"])
plt.grid(axis="y", linestyle="--", alpha=0.7)  # adds grid for neatness
plt.show()
```

```python
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]

plt.scatter(
    x, 
    y, 
    c=y,                  
    cmap="viridis",       
    marker="o",           
    s=100,                
    edgecolor="black"     
)

plt.title("Scatter Plot Example", fontsize=16, fontweight="bold")
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.colorbar(label="Y-value")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
data = np.random.normal(100, 20, 200)

plt.figure(figsize=(8,6))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot Example")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
langs = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
explode = (0.1, 0, 0, 0, 0, 0)

plt.pie(x, labels=langs, explode=explode, autopct="%1.1f%%", startangle=90, shadow=True)
plt.title("Pie Chart Example")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 1, 2, 1, 2, 1]
y_error = 0.2

plt.plot(x, y, linestyle="--", color="blue", alpha=0.7)
plt.errorbar(x, y, yerr=y_error, fmt="o", color="red", ecolor="black", capsize=5)
plt.title("Error Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot3D(x, y, z, color="green", marker="o")
ax.set_title("3D Line Plot Example")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker="o", color="orange")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Line Plot with Labels")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, label="Line 1", marker="o")
plt.plot(x, y2, label="Line 2", marker="s")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Line Plot Example")
plt.legend()
plt.show()
```
