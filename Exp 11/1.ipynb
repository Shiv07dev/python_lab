```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg")
img_array = np.array(img)

# Crop coordinates (Top-left to Bottom-right)
y1, x1 = 100, 100
y2, x2 = 250, 200
cropped_img = img_array[y1:y2, x1:x2]

# Show images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cropped_img)
plt.title('Cropped Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg")
img_array = np.array(img)

# Create negative
negative_img = 255 - img_array

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(negative_img)
plt.title('Negative Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load in grayscale
img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg").convert("L")
img_array = np.array(img)

# Thresholding
threshold = 128
binary_img = np.where(img_array < threshold, 0, 255).astype(np.uint8)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.title('Binarized Image (Threshold=128)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg")
img_array = np.array(img)

# Grayscale conversion formula
gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array)
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load grayscale
img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg").convert("L")
img_array = np.array(img)

# Histogram
plt.figure(figsize=(10, 5))
plt.hist(img_array.flatten(), bins=256, range=(0, 256), color='gray')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Pixel Intensity Histogram')
plt.grid(True)
plt.show()
```
