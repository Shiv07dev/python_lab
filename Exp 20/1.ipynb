```python

import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf

# Load again (optional)
audio, sample_rate = sf.read(r"C:\Users\shivk\Downloads\9df1c550ea5744e3a4fe87816570f260.mp3")

# Create time axis (seconds)
time = np.arange(0, len(audio)) / sample_rate

# Plot waveform
plt.figure(figsize=(10,4))
plt.plot(time, audio, color='purple')
plt.title("Audio Signal in Time Domain")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
from pydub import AudioSegment

# Load audio (Pydub works with file paths)
audio = AudioSegment.from_file(r"C:\Users\shivk\Downloads\9df1c550ea5744e3a4fe87816570f260.mp3")

# Apply 2-second fade-in
audio_fade_in = audio.fade_in(2000)

# Save result
audio_fade_in.export("audio_file_fade_in.wav", format="wav")

print("âœ¨ Fade-in effect applied and saved as audio_file_fade_in.wav")
```
