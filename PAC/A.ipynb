```python
# Program to merge two lists element-wise and replace values greater than 10 with "ICT"

# Taking input for both lists
list1 = list(map(int, input("Enter numbers for first list (space separated): ").split()))
list2 = list(map(int, input("Enter numbers for second list (space separated): ").split()))

merged = []  # empty list for result

for i in range(len(list1)):
    total = list1[i] + list2[i]
    if total > 10:
        merged.append("ICT")
    else:
        merged.append(total)

print("Merged list:", merged)
```
